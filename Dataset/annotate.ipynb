```python
import yaml
import re

foodtype = [
    r'((G|g)riech(\w*))', 
    r'((C|c)hines(\w*))', 
    r'((I|i)talien(\w*))', 
    r'((F|f)ranzös(\w*))',
    r'((D|d)eutsch(\w*))', 
    r'((A|a)siat(\w*))', 
    r'((S|s)izilian(\w*))', 
    r'((A|a)rgentin(\w*))', 
    r'((M|m)exikan(\w)*)', 
    r'((M|m)editeran(\w)*)', 
    r'((K|k)orean(\w)*)', 
    r'((I|i)ndisch(\w)*)', 
    r'(Ente)', 
    r'((P|p)ute(\w*))', 
    r'((G|g)eflügel(\w*))', 
    r'((H|h)ausmannskost)']
locations = [
    r'(Weitmar)',
    r'(Innenstadt)',
    r'(Querenburg)',
    r'(Linden)',
    r'(Grumme)',
    r'(Hamme)',
    r'(Werne)',
    r'(Wiemelhausen)',
    r'(Langendreer)',
    r'(Zentrum)',
    r'(Bochum)']
dates = [
    r'((M|m)ontag(\w*))',
    r'((D|d)ienstag(\w*))',
    r'((M|m)ittwoch(\w*))',
    r'((D|d)onnerstag(\w*))',
    r'((F|f)reitag(\w*))',
    r'((S|s)amstag(\w*))',
    r'((S|s)onntag(\w*))',
]
fields = [
    r'(Ort)',
    r'(Öffnungszeiten)',
    r'(Art der Küche)',
    r'(Art des Restaurants)',
    r'(Preis der Mahlzeit)',
    r'(Tageszeit)']

with open('germanNLU.yml', 'r') as file:
    data = yaml.safe_load(file)

#print(data['nlu'][0]['example'])

inputProvide = data['nlu'][0]['example'].strip('- ').strip('\n').split('\n- ')

outputProvide = []
for word in inputProvide:
    counter = 0
    for regex in foodtype:
        result = re.sub(regex, r"[\1](foodtype)", word)
        if result != word:
            outputProvide.append(result)
            break
        elif counter == 15:
            outputProvide.append(word)
        else: 
            counter += 1

outputProvide2 = []
for word in outputProvide:
    counter = 0
    for regex in locations:        
        result = re.sub(regex, r"[\1](location)", word)
        if result != word:
            outputProvide2.append(result)
            break
        elif counter == 10:
            outputProvide2.append(word)
        else: 
            counter += 1

outputProvide3 = []
for word in outputProvide2:
    counter = 0
    for regex in dates:        
        result = re.sub(regex, r"[\1](date)", word)
        if result != word:
            outputProvide3.append(result)
            break
        elif counter == 6:
            outputProvide3.append(word)
        else: 
            counter += 1
print(len(inputProvide))
print(len(outputProvide))
print(len(outputProvide2))
print(len(outputProvide3))

print(outputProvide3)
```
