```python
import yaml
import re

regexDict = {
    'foodtype': [
    r'((G|g)riech(\w*))', 
    r'((C|c)hines(\w*))', 
    r'((I|i)talien(\w*))', 
    r'((F|f)ranzös(\w*))',
    r'((D|d)eutsch(\w*))', 
    r'((A|a)siat(\w*))', 
    r'((S|s)izilian(\w*))', 
    r'((A|a)rgentin(\w*))', 
    r'((M|m)exikan(\w)*)', 
    r'((M|m)editeran(\w)*)',
    r'((K|k)orean(\w)*)', 
    r'((I|i)ndisch(\w)*)', 
    r'(Ente)', 
    r'((P|p)ute(\w*))', 
    r'((G|g)eflügel(\w*))', 
    r'((H|h)ausmannskost)'],

    "location": [
    r'(Weitmar)',
    r'(Innenstadt)',
    r'(Querenburg)',
    r'(Linden)',
    r'(Grumme)',
    r'(Hamme)',
    r'(Werne)',
    r'(Wiemelhausen)',
    r'(Langendreer)',
    r'(Zentrum)',
    r'(Bochum)'],

    "date":[
    r'((M|m)ontag(\w*))',
    r'((D|d)ienstag(\w*))',
    r'((M|m)ittwoch(\w*))',
    r'((D|d)onnerstag(\w*))',
    r'((F|f)reitag(\w*))',
    r'((S|s)amstag(\w*))',
    r'((S|s)onntag(\w*))',],

    "field": [
    r'(Ort)',
    r'(Öffnungszeiten)',
    r'(Art der Küche)',
    r'(Art des Restaurants)',
    r'(Preis der Mahlzeit)',
    r'(Tageszeit)'],

    "time":[
    r'((A|a)bend(\w*))',
    r'((M|m)ittag(\w*))',
    r'((M|m)orgens)',],

    "price":[
    r'([0-9]* (E|e)uro(\w*))',
    r'((\w*) (P|p)reis(\w*))',],   
    
    "logical":[
    r'((J|j)a)',
    r'((N|n)ein)',]
    }

#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

def appendDataset(key, currIdx, input):
    #print(currIdx)
    output = []
    for word in input:
        counter = 0
        for regex in regexDict[key]:
            result = re.sub(regex, r"[\1]("+key+")", word)
            if result != word:
                output.append(result)
                break
            elif counter == (len(regexDict[key])-1):
                output.append(word)
            else: 
                counter += 1
    if currIdx + 1 == 7:
        return output

    return appendDataset(list(regexDict.keys())[currIdx +1], currIdx + 1, output)

#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

with open('germanNLU.yml', 'r') as file:
    data = yaml.safe_load(file)

finalString = ''

for intents in range(8):
    inputs = data['nlu'][intents]['example'].strip('- ').strip('\n').split('\n- ')
    finalOutput = appendDataset(list(regexDict.keys())[0] ,0, inputs)
    for word in finalOutput:
        finalString += "    - "+word+"\n"
    finalString += "\n \n"
print(finalString)
```
